```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,2,3,4]
y = [10,20,15,25]
plt.plot(x,y)
plt.show()
```

```python
x=['Mon','Tue','Wed','Thu','Fri']
y = [10,20,15,25,35]
plt.plot(x,y,color='blue',linestyle='dashed',linewidth=3,marker='o')
plt.title('Daily sales in this week')
plt.xlabel('Days in week')
plt.ylabel('Sales in Dkk')
plt.legend(['sales in week'],loc='lower right',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.show()
```
